```python
data = pd.read_csv('parks.csv')
print(data)
```

```python
x = np.array(data[['Longitude']])
y = np.array(data[['Latitude']])

plt.figure()
plt.scatter(x,y)
plt.title('Park Location')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
labels = np.array(data[['Name']])

fig, ax = plt.subplots()
ax.scatter(x,y)

for i, txt in enumerate(labels):
    ax.annotate(txt,(x[i],y[i]))
```
